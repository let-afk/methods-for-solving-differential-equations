```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
def func_a(x):
    return x ** 2

def func_b(x):
    return np.exp(x)

def func_c(x):
    return 5 + x ** 3 + np.sin(x)


def draw(file, h, f, x_lim_start = 0, x_lim_end = 10, y_lim_start = 0, y_lim_end = 100):
    df = pd.read_csv(f"{file}", sep=";")
    fig, ax = plt.subplots(figsize=(6, 3))

    ax.scatter(x=df["t"], y=df["y"], label='approximate solution')

    t = np.linspace(0, 10, 5000)
    y = f(t)
    ax.plot(t, y, color='red', label='exact solution')
    
    ax.set_xlim(x_lim_start, x_lim_end)
    ax.set_ylim(y_lim_start, y_lim_end)
    ax.legend()
```

```python
draw('task4-h1.csv', 1, func_a), draw('task4-h01.csv', 0.1, func_a, 8, 10, 64, 100), draw('task4-h001.csv', 0.01, func_a, 9.5, 10, 90.25, 100)
```

```python
draw('task4-b-h1.csv', 1, func_b, 0, 10, 0, 10000), draw('task4-b-h01.csv', 0.1, func_b, 0, 10, 0, 10000), draw('task4-b-h001.csv', 0.01, func_b, 0, 10, 0, 10000)
```

```python
draw('task4-c-h1.csv', 1, func_c, 0, 10, 0, 1000), draw('task4-c-h01.csv', 0.1, func_c, 0, 10, 0, 1000), draw('task4-c-h001.csv', 0.01, func_c, 0, 10, 0, 1000)
```
