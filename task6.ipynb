```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
def draw(file, h, additional_point_in_graph=False):
    df = pd.read_csv(f"{file}", sep=";")
    fig, ax = plt.subplots(figsize=(6, 3))

    ax.plot(df["x"], df["z"])
    
    if additional_point_in_graph:
        ax.scatter(df["x"], df["z"])
    
    # ax.set_xlim(x_lim_start, x_lim_end)
    # ax.set_ylim(z_lim_start, z_lim_end)
    ax.set_title(f"h = {h}")
    plt.xlabel("x")
    plt.ylabel("z")
    ax.legend()
```

```python
draw('task6-h1.csv', 1, True), draw('task6-h01.csv', 0.1), draw('task6-h001.csv', 0.01)
```
