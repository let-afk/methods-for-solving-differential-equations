```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
def draw(file, h, additional_point_in_graph=False):
    df = pd.read_csv(f"{file}", sep=";")
    fig, ax = plt.subplots(figsize=(6, 3))
    
    fig.set_figheight(7)
    fig.set_figwidth(7)

    ax.plot(df["x"], df["z"])
    
    if additional_point_in_graph:
        ax.scatter(df["x"], df["z"])
    
    # ax.set_xlim(x_lim_start, x_lim_end)
    # ax.set_ylim(z_lim_start, z_lim_end)
    ax.set_title(f"h = {h}")
    plt.xlabel("x")
    plt.ylabel("z")
    ax.legend()
```

```python
draw('task7-h01.csv', 0.1)
```

```python
draw('task7-h01.csv', 0.1)
```

```python
def draw_error(file, h, additional_point_in_graph=False):
    df = pd.read_csv(f"{file}", sep=";")
    df = df.iloc[0:300]
    fig, ax = plt.subplots(figsize=(6, 3))
    
    fig.set_figheight(7)
    fig.set_figwidth(7)

    ax.plot(df["count"], df["error"])
    
    if additional_point_in_graph:
        ax.scatter(df["count"], df["error"])
    
    # ax.set_xlim(x_lim_start, x_lim_end)
    # ax.set_ylim(z_lim_start, z_lim_end)
    ax.set_title(f"h = {h}")
    plt.xlabel("count")
    plt.ylabel("error")
    ax.legend()

draw_error('task7-error.csv', 0.1)
```
